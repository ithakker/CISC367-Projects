```python
# Try Yourself Problem 1: 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
df = sns.load_dataset('titanic')

sns.barplot(x = 'who',
            y = 'fare',
            hue = 'class',
            data = df,
            estimator = np.median
            )
plt.show()
```

```python
# Try Yourself Problem 2
titanic_dataset = sns.load_dataset("titanic")
sns.barplot(x = "sex", y = "survived", data = titanic_dataset)
plt.show()
```

```python
# Try yourself Problem 3
xData = ['A', 'B', 'C']
yData = [1, 5, 3]

sns.barplot(x = xData, y = yData, palette = "Blues")
plt.show()
```

```python
# Try Yourself problem 4
df = sns.load_dataset('titanic')
sns.barplot(x = 'fare',
            y = 'who',
            hue = 'embark_town',
            data = df,
            palette = "Greens")
plt.show()
```

```python
# Try Yourself Problem 5
df = sns.load_dataset('titanic')
sns.barplot(x = 'class',
            y = 'survived',
            hue = 'sex',
            order = ["Third", "Second", "First"],
            data = df,
            palette = "Reds",
           ci = 0)
plt.show()
```
